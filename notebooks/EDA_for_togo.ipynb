```python
# Check for negative or zero values where only positive should exist
invalid_values_ghi = df[df['GHI'] <= 0]
invalid_values_dni = df[df['DNI'] <= 0]
invalid_values_dhi = df[df['DHI'] <= 0]
invalid_values_moda = df[df['ModA'] <= 0]
invalid_values_modb = df[df['ModB'] <= 0]

print("\nInvalid GHI Values:\n", invalid_values_ghi)
print("\nInvalid DNI Values:\n", invalid_values_dni)
print("\nInvalid DHI Values:\n", invalid_values_dhi)
print("\nInvalid ModA Values:\n", invalid_values_moda)
print("\nInvalid ModB Values:\n", invalid_values_modb)
```

```python
def detect_and_replace_negative_outliers_iqr(data):
    Q1 = data.quantile(0.25)
    Q3 = data.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    
    # Identify negative outliers
    outliers = data[data < lower_bound]
    
    # Replace negative outliers with zero
    data[data < lower_bound] = 0
    
    return data

# Replace negative outliers in the 'GHI' column
df['GHI'] = detect_and_replace_negative_outliers_iqr(df['GHI'])

print("Updated GHI column with negative outliers replaced by zero:\n", df['GHI'])
```

```python
# Replace all negative values in the 'GHI' column with zero
df['GHI'] = df['GHI'].apply(lambda x: max(x, 0))

print("Updated GHI column with negative values replaced by zero:\n", df['GHI'])
```

```python
df['Timestamp'] = pd.to_datetime(df['Timestamp'])
df.set_index('Timestamp', inplace=True)
```

```python
#Univariant Analysis
# univariate plots
# Plotting GHI, DNI, DHI, and Tamb over time
plt.figure(figsize=(14, 8))

# Line plots for each parameter
plt.subplot(2, 2, 1)
plt.plot(df.index, df['GHI'], label='GHI', color='orange')
plt.title('Global Horizontal Irradiance (GHI) Over Time')
plt.xlabel('Time')
plt.ylabel('GHI')
plt.grid(True)

plt.subplot(2, 2, 2)
plt.plot(df.index, df['DNI'], label='DNI', color='blue')
plt.title('Direct Normal Irradiance (DNI) Over Time')
plt.xlabel('Time')
plt.ylabel('DNI')
plt.grid(True)

plt.subplot(2, 2, 3)
plt.plot(df.index, df['DHI'], label='DHI', color='green')
plt.title('Diffuse Horizontal Irradiance (DHI) Over Time')
plt.xlabel('Time')
plt.ylabel('DHI')
plt.grid(True)

plt.subplot(2, 2, 4)
plt.plot(df.index, df['Tamb'], label='Tamb', color='red')
plt.title('Ambient Temperature (Tamb) Over Time')
plt.xlabel('Time')
plt.ylabel('Temperature (°C)')
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
# Create a new column for the cleaned and uncleaned state
df['Cleaning_Flag'] = df['Cleaning'].apply(lambda x: 'Cleaned' if x == 'Yes' else 'Uncleaned')

# Plotting ModA and ModB over time, with cleaning flag
plt.figure(figsize=(14, 6))

# ModA Plot
plt.subplot(1, 2, 1)
sns.lineplot(data=df, x=df.index, y='ModA', hue='Cleaning_Flag')
plt.title('ModA Sensor Readings Over Time')
plt.xlabel('Time')
plt.ylabel('ModA')
plt.grid(True)

# ModB Plot
plt.subplot(1, 2, 2)
sns.lineplot(data=df, x=df.index, y='ModB', hue='Cleaning_Flag')
plt.title('ModB Sensor Readings Over Time')
plt.xlabel('Time')
plt.ylabel('ModB')
plt.grid(True)

plt.tight_layout()
plt.show()
```
